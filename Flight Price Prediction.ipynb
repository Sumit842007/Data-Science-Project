```python
x['Date_of_Journey']=pd.to_datetime(x['Date_of_Journey'])
```

```python
x["Journey_day"] = pd.to_datetime(x.Date_of_Journey, format="%d/%m/%Y").dt.day
x["Journey_month"] = pd.to_datetime(x["Date_of_Journey"], format = "%d/%m/%Y").dt.month
x.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
x["Dep_hour"] = pd.to_datetime(x["Dep_Time"]).dt.hour
x["Dep_min"] = pd.to_datetime(x["Dep_Time"]).dt.minute
x.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
x["Arrival_hour"] = pd.to_datetime(x.Arrival_Time).dt.hour
x["Arrival_min"] = pd.to_datetime(x.Arrival_Time).dt.minute
x.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
def get_duration(x1):
    x1=x1.split(' ')
    hours=0
    mins=0
    if len(x1)==1:
        x1=x1[0]
        if x1[-1]=='h':
            hours=int(x1[:-1])
        else:
            mins=int(x1[:-1])
    else:
        hours=int(x1[0][:-1])
        mins=int(x1[1][:-1])
    return hours,mins
x['Duration_hours']=x.Duration.apply(lambda x1:get_duration(x1)[0])
x['Duration_mins']=x.Duration.apply(lambda x1:get_duration(x1)[1])
x.drop(["Duration"], axis = 1, inplace = True)
```

```python
Y=df_train['Price']
```

```python
for i in range(0,1000):
    features_train,features_test,target_train,target_test=train_test_split(x_over,Y,test_size=0.2,random_state=i)
    lr.fit(features_train,target_train)
    pred_train=lr.predict(features_train)
    pred_test=lr.predict(features_test)

    print(f"At random state {i},the training accuracy is:-{r2_score(target_train,pred_train)}")
    print(f"At random state {i},the training accuracy is:-{r2_score(target_test,pred_test)}")
    print("\n")
```
